```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mping
import argparse
```

```python
df = pd.read_excel('Mishawaka Scout All Matches 2.xlsm')
heat = df.iloc[:,[13]]
heat_temp = heat
heat_temp = heat_temp.dropna()
heat_temp.isnull()
print(heat_temp)
```

```python
hs = heat_temp.convert_dtypes('str')
r = 0
vals = []
while r < len(hs):
    pin = hs.iat[r, 0]
    vals.append(pin)
    r = r + 1

all = []
hmi = [eval(v) for v in vals]
for h in hmi:
    parse = h
    for p in parse:
        all.append(p)

print(all)
```

```python
x_val = []
y_val = []
for a in all:
    #y values :D
    if (28 <= a <= 36):
        y = .5
        y_val.append(y)  
    elif (19 <= a <= 27):
        y = 1.5
        y_val.append(y)
    elif (10 <= a <= 18):
        y = 2.5
        y_val.append(y) 
    elif (1 <= a <= 9):
        y = 3.5
        y_val.append(y)

    #x vals
    if (a == 1 or a == 10 or a == 19 or a == 28):
        x = .5
        x_val.append(x)
    elif (a == 2 or a == 11 or a == 20 or a == 29):
        x = 1.5
        x_val.append(x)
    elif (a == 3 or a == 12 or a == 21 or a == 30):
        x = 2.5
        x_val.append(x)
    elif (a == 4 or a == 13 or a == 22 or a == 31):
        x = 3.5
        x_val.append(x)
    elif (a == 5 or a == 14 or a == 23 or a == 32):
        x = 4.5
        x_val.append(x)
    elif (a == 6 or a == 15 or a == 24 or a == 33):
        x = 5.5
        x_val.append(x)
    elif (a == 7 or a == 16 or a == 25 or a == 34):
        x = 6.5
        x_val.append(x)
    elif (a == 8 or a == 17 or a == 26 or a == 35):
        x = 7.5
        x_val.append(x)
    elif (a == 9 or a == 18 or a == 27 or a == 36):
        x = 8.5
        x_val.append(x)
```

```python
img = mping.imread('gridimg.png')
fig, ax = plt.subplots()
ax.imshow(img, extent = [0, 9, 0, 4])

col = np.random.rand(len(x_val),3)
color = (0.0, 1.0, 1.0) * col

ax.scatter(x_val, y_val, 200, c=color, alpha=0.05)
ax.set(xlim=(0,9), xticks=np.arange(0,10), ylim=(0,4), yticks=np.arange(0,5))
ax.set_xlabel("x coords")
ax.set_ylabel("y coords")
ax.legend()
```
