```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
die = pd.DataFrame([1,2,3,4,5,6])
sumOfDice = die.sample(2, replace = True).sum().loc[0]
print("Sum of dice is ", sumOfDice)
```

```python
trial = 10000
result = []
for i in range(trial):
    j = sumOfDice = die.sample(2, replace = True).sum().loc[0]
    result.append(j)
result = np.array(result)
result
```

```python
freq = pd.DataFrame(result)[0].value_counts()
sortFreq = freq.sort_index()
sortFreq
```

```python
sortFreq.plot(kind='bar',color='purple')
```

```python
relFreq = sortFreq / trial
relFreq.plot(kind='bar',color='purple')
```

```python
sns.distplot(relFreq)
```
